```python
df = df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type',
                             'targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
```

```python
df = df[['Year','Month','Day','Country','Region','State','City','Attack_type','Target_type','Target_subtype','Gang_name','Weapon_type']]
```

```python
df
```

```python
df['Target_subtype'] = df['Target_subtype'].fillna('others')
```

```python
df['Target_type'] = df['Target_type'].fillna('others')
```

```python
df['Weapon_type'] = df['Weapon_type'].fillna('others')
```

```python
df['Attack_type'] = df['Attack_type'].fillna('others')
```

```python
df['State'] = df['State'].fillna('Unknown')
```

```python
df['Gang_name'] = df['Gang_name'].fillna('Unknown')
```

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
df.replace( {"Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)": "Vehicle"},inplace = True)
```

# **Visulization**

```python
plt.figure(figsize = (20,5));
sns.countplot(x = df.Weapon_type, palette='viridis')
```

```python
plt.figure(figsize = (20,5));
plt.xticks(rotation = 90)
plt.title('Attacks by Year')
sns.countplot(x = df.Year,palette='viridis')
```

```python
plt.figure(figsize = (20,5));
# plt.xticks(rotation = 90) # if uncomment this part then 1 2 3 4 5...... is rotate by 90 as see above part
plt.title('Attacks by Month')
sns.countplot(x = df.Month,palette='viridis')
```

```python
plt.figure(figsize = (20,5));
plt.title('Attacks by Day')
sns.countplot(x = df.Day,palette='viridis')
```

```python
plt.figure(figsize = (30,5));
plt.rcParams.update({'font.size' : 8})
plt.xticks(rotation = 90)
plt.title('Attacks by Country')
sns.countplot(x = df.Country,palette='viridis')
```

```python
plt.figure(figsize=(30,5))
plt.xticks(rotation = 0)
plt.rcParams.update({'font.size':10})
plt.title('Attacks by region')
sns.countplot(x = df.Region, palette = 'viridis')
```

```python
plt.figure(figsize = (20,5))
plt.xticks(rotation = 90)
plt.rcParams.update({'font.size' : 10})
plt.title('Based on Target type')
sns.countplot(x = df.Target_type, palette = 'viridis')
```
